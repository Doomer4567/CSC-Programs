```python
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl
from sklearn.datasets import make_circles
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
X_train = data.data[0:20,:2]
Y_train = data.target[0:20]
X_test = data.data[21:31,:2]
Y_test = data.target[21:31]

plt.figure(figsize=(8, 6))

for label in [0, 1]:
    plt.scatter(
        X_train[Y_train == label, 0],
        X_train[Y_train == label, 1],
        label=f'Train Class {label}',
        marker='o',
        edgecolor='black'
    )

for label in [0, 1]:
    plt.scatter(
        X_test[Y_test == label, 0], 
        X_test[Y_test == label, 1],
        label=f'Test Class {label}',
        marker='x'
    )

# Labels and legend
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.title('Breast Cancer Data (First 2 Features)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Start of the Assignment


```python
from qiskit.circuit.library import ZZFeatureMap
from qiskit_machine_learning.kernels import FidelityQuantumKernel
from sklearn.svm import SVC

#imports yucky

#Creating feature maps for the data
adhoc_feature_map = ZZFeatureMap(feature_dimension=2, reps=2, entanglement='linear')
adhoc_kernel = FidelityQuantumKernel(feature_map=adhoc_feature_map)

#SVC algorithm thingy
model2 = SVC(kernel=adhoc_kernel.evaluate)
model2.fit(X_train,Y_train)
Model2_score = model2.score(X_test,Y_test)

print("Score: " + str(Model2_score))
#working simulation Horray
```

```python
#Let's do it again

Feature_Map = ZZFeatureMap(feature_dimension=2, reps=3, entanglement='linear')
Kernal = FidelityQuantumKernel(feature_map=Feature_Map)

Model = SVC(kernel=Kernal.evaluate)
Model.fit(X_train,Y_train)

Model_score = Model.score(X_test,Y_test)

print("Quantum SVM Score: " + str(Model_score))

Classic_Model = SVC()
Classic_Model.fit(X_train,Y_train)

Classic_Model_Score = Classic_Model.score(X_test,Y_test)

print("Classical SVC Score: " + str(Classic_Model_Score))
```
